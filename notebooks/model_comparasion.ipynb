# Model comparison

*Objetivo:*
El objetivo es comparar diferentes modelos de Machine Learning para predecir la intención de compra de usuarios en un ecommerce a partir de su comportamiento de navegación. En este caso Logistic Regression con el modelo de RandomForest hecho ne (purchase_prediction_model)

```python
#Importación librerías

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler
)

from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score
)

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
```

```python
#Cargar dataset

df = pd.read_csv("../data/raw/data.csv")

df.head()
```

```python
#Muestra la info

df.info()
```

```python
#Muestra si hay nulos

df.isnull().sum()
```

```python
#Comprobar duplicados

df.duplicated().sum()
```

```python
#Eliminar duplicados

df = df.drop_duplicates()
```

```python
#Comprobar eliminación de duplicados

df.duplicated().sum()
```

```python
#Convertir target

df["Revenue"] = df["Revenue"].astype(int)
```

```python
display(df)
```

```python
#Separar X e Y

X = df.drop("Revenue", axis=1)
y = df["Revenue"]
```

```python
#Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,random_state=42)
```

```python
#separar columnas categóricas y numéricas

cat_cols = ["Month", "VisitorType", "Weekend"]

num_cols = [col for col in X.columns if col not in cat_cols]
```

```python
#Preprocessing

preprocessor = ColumnTransformer([("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),("num", StandardScaler(), num_cols)])
```

```python
#Pipeline

logistic_pipeline = ImbPipeline([("preprocessing", preprocessor), ("smote", SMOTE(random_state=42)), ("model", LogisticRegression(max_iter=1000, random_state=42))])
```

```python
#Entrenar

logistic_pipeline.fit(X_train, y_train)
```

```python
#Predicciones

y_pred = logistic_pipeline.predict(X_test)

y_proba = logistic_pipeline.predict_proba(X_test)[:,1]
```

```python
#Coeficientes

coefficients = logistic_pipeline.named_steps["model"].coef_[0]
```

```python
#Evaluación 1

print(classification_report(y_test, y_pred))
```

```python
#Evaluación 2

print(confusion_matrix(y_test, y_pred))
```

```python
#Evaluación 3

print("ROC-AUC:", roc_auc_score(y_test, y_proba))
```

```python
#Métricas

classification_report
confusion_matrix
roc_auc_score
```
